Question 2

Write a function that takes a list sorted in non-decreasing order and deletes any duplicate nodes from the list. The list should only be traversed once.

For example if the linked list is 11->11->11->21->43->43->60 then removeDuplicates() should convert the list to 11->21->43->60.

Input:
LinkedList: 
11->11->11->21->43->43->60
Output:
11->21->43->60

Input:
LinkedList: 
10->12->12->25->25->25->34
Output:
10->12->25->34

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeDuplicates(head):
    if not head:
        return head

    current = head
    while current.next:
        if current.val == current.next.val:
            current.next = current.next.next
        else:
            current = current.next

    return head

# Example 1:
head1 = ListNode(11)
head1.next = ListNode(11)
head1.next.next = ListNode(11)
head1.next.next.next = ListNode(21)
head1.next.next.next.next = ListNode(43)
head1.next.next.next.next.next = ListNode(43)
head1.next.next.next.next.next.next = ListNode(60)

newHead1 = removeDuplicates(head1)

# Print the modified linked list
while newHead1:
    print(newHead1.val, end=" ")
    newHead1 = newHead1.next
# Output: 11 21 43 60

# Example 2:
head2 = ListNode(10)
head2.next = ListNode(12)
head2.next.next = ListNode(12)
head2.next.next.next = ListNode(25)
head2.next.next.next.next = ListNode(25)
head2.next.next.next.next.next = ListNode(25)
head2.next.next.next.next.next.next = ListNode(34)

newHead2 = removeDuplicates(head2)

# Print the modified linked list
while newHead2:
    print(newHead2.val, end=" ")
    newHead2 = newHead2.next
# Output: 10 12 25 34
```

Input:
LinkedList: 1->2->2->4->5->6->7->8
K = 4
Output:4 2 2 1 8 7 6 5
Explanation:
The first 4 elements 1,2,2,4 are reversed first
and then the next 4 elements 5,6,7,8. Hence, the
resultant linked list is 4->2->2->1->8->7->6->5.

Input:
LinkedList: 1->2->3->4->5
K = 3
Output:3 2 1 5 4
Explanation:
The first 3 elements are 1,2,3 are reversed
first and then elements 4,5 are reversed.Hence,
the resultant linked list is 3->2->1->5->4.

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseKNodes(head, k):
    if not head or k == 1:
        return head

    dummy = ListNode(0)
    dummy.next = head
    prev = dummy

    while True:
        count = 0
        curr = prev.next
        while curr:
            count += 1
            curr = curr.next
            if count == k:
                break

        if count == k:
            next_group = curr.next
            curr.next = None
            prev.next = reverseLinkedList(prev.next)
            curr = prev.next

            while curr.next:
                curr = curr.next

            curr.next = next_group
            prev = curr
        else:
            break

    return dummy.next

def reverseLinkedList(head):
    prev = None
    curr = head

    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node

    return prev

# Example 1:
head1 = ListNode(1)
head1.next = ListNode(2)
head1.next.next = ListNode(2)
head1.next.next.next = ListNode(4)
head1.next.next.next.next = ListNode(5)
head1.next.next.next.next.next = ListNode(6)
head1.next.next.next.next.next.next = ListNode(7)
head1.next.next.next.next.next.next.next = ListNode(8)

k1 = 4

newHead1 = reverseKNodes(head1, k1)

# Print the modified linked list
while newHead1:
    print(newHead1.val, end=" ")
    newHead1 = newHead1.next
# Output: 4 2 2 1 8 7 6 5

# Example 2:
head2 = ListNode(1)
head2.next = ListNode(2)
head2.next.next = ListNode(3)
head2.next.next.next = ListNode(4)
head2.next.next.next.next = ListNode(5)

k2 = 3

newHead2 = reverseKNodes(head2, k2)

# Print the modified linked list
while newHead2:
    print(newHead2.val, end=" ")
    newHead2 = newHead2.next
# Output: 3 2 1 5 4
```

Question 4

Given a linked list, write a function to reverse every alternate k nodes (where k is an input to the function) in an efficient way. Give the complexity of your algorithm.

Inputs:   1->2->3->4->5->6->7->8->9->NULL and k = 3
Output:   3->2->1->4->5->6->9->8->7->NULL.

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseAlternateKNodes(head, k):
    if not head or k == 1:
        return head

    dummy = ListNode(0)
    dummy.next = head
    prev = dummy
    curr = head
    is_reversed = False

    while curr:
        count = 1
        while curr and count < k:
            curr = curr.next
            count += 1

        if curr:
            next_group = curr.next
            curr.next = None

            if is_reversed:
                prev.next = reverseLinkedList(prev.next)
                prev = prev.next
            else:
                prev = curr

            curr = next_group
            is_reversed = not is_reversed

    return dummy.next

def reverseLinkedList(head):
    prev = None
    curr = head

    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node

    return prev

head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
head.next.next.next.next.next = ListNode(6)
head.next.next.next.next.next.next = ListNode(7)
head.next.next.next.next.next.next.next = ListNode(8)
head.next.next.next.next.next.next.next.next = ListNode(9)

k = 3

newHead = reverseAlternateKNodes(head, k)

# Print the modified linked list
while newHead:
    print(newHead.val, end=" ")
    newHead = newHead.next
# Output: 3 2 1 4 5 6 9 8 7
```

Question 5

Given a linked list and a key to be deleted. Delete last occurrence of key from linked. The list may have duplicates.
Input:   1->2->3->5->2->10, key = 2
Output:  1->2->3->5->10

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def deleteLastOccurrence(head, key):
    dummy = ListNode(0)
    dummy.next = head
    prev = None
    last_occurrence = None
    curr = dummy

    # Find the last occurrence of the key and its previous node
    while curr.next:
        if curr.next.val == key:
            prev = curr
            last_occurrence = curr.next
        curr = curr.next

    if last_occurrence:
        # Delete the last occurrence of the key
        prev.next = last_occurrence.next
        last_occurrence.next = None

        # Handle the case where the last occurrence is the head of the linked list
        if last_occurrence == head:
            head = prev.next

    return head

head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(5)
head.next.next.next.next = ListNode(2)
head.next.next.next.next.next = ListNode(10)

key = 2

newHead = deleteLastOccurrence(head, key)

# Print the modified linked list
while newHead:
    print(newHead.val, end=" ")
    newHead = newHead.next
# Output: 1 2 3 5 10
```

Question 6

Given two sorted linked lists consisting of N and M nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.
Examples:

Input: a: 5->10->15, b: 2->3->20

Output: 2->3->5->10->15->20

Input: a: 1->1, b: 2->4

Output: 1->1->2->4

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def mergeLists(a, b):
    dummy = ListNode(0)
    prev = dummy
    currA = a
    currB = b

    while currA and currB:
        if currA.val <= currB.val:
            prev.next = currA
            currA = currA.next
        else:
            prev.next = currB
            currB = currB.next
        prev = prev.next

    # Append remaining nodes from list a or b
    if currA:
        prev.next = currA
    if currB:
        prev.next = currB

    return dummy.next

# Linked list a: 5->10->15
a = ListNode(5)
a.next = ListNode(10)
a.next.next = ListNode(15)

# Linked list b: 2->3->20
b = ListNode(2)
b.next = ListNode(3)
b.next.next = ListNode(20)

mergedHead = mergeLists(a, b)

# Print the merged list
while mergedHead:
    print(mergedHead.val, end=" ")
    mergedHead = mergedHead.next
# Output: 2 3 5 10 15 20
```

Question 7

Given a Doubly Linked List, the task is to reverse the given Doubly Linked List.

Example:
Original Linked list 10 8 4 2
Reversed Linked list 2 4 8 10

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def reverseDoublyLinkedList(head):
    if head is None or head.next is None:
        return head

    prev = None
    curr = head

    while curr:
        next_node = curr.next
        curr.next = prev
        curr.prev = next_node
        prev = curr
        curr = next_node

    head = prev
    return head


# Create a doubly linked list
head = Node(10)
head.next = Node(8)
head.next.prev = head
head.next.next = Node(4)
head.next.next.prev = head.next
head.next.next.next = Node(2)
head.next.next.next.prev = head.next.next

# Reverse the doubly linked list
reversed_head = reverseDoublyLinkedList(head)

# Print the reversed list
while reversed_head:
    print(reversed_head.data, end=" ")
    reversed_head = reversed_head.next
# Output: 2 4 8 10
```

Question 8

Given a doubly linked list and a position. The task is to delete a node from given position in a doubly linked list.

Input:
LinkedList = 1 <--> 3 <--> 4
x = 3
Output:1 3
Explanation:After deleting the node at
position 3 (position starts from 1),
the linked list will be now as 1->3.

Input:
LinkedList = 1 <--> 5 <--> 2 <--> 9
x = 1
Output:5 2 9

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def deleteNodeFromPosition(head, position):
    if position == 1:
        new_head = head.next
        if new_head:
            new_head.prev = None
        return new_head

    current = head
    for _ in range(position - 1):
        current = current.next

    prev_node = current.prev
    next_node = current.next

    prev_node.next = next_node
    if next_node:
        next_node.prev = prev_node

    return head


# Create a doubly linked list
head = Node(1)
head.next = Node(5)
head.next.prev = head
head.next.next = Node(2)
head.next.next.prev = head.next
head.next.next.next = Node(9)
head.next.next.next.prev = head.next.next

# Delete a node from a given position
position = 1
updated_head = deleteNodeFromPosition(head, position)

# Print the updated list
current = updated_head
while current:
    print(current.data, end=" ")
    current = current.next
# Output: 5 2 9
```
